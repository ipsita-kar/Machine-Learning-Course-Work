# Using the data ‘football.csv’ do the following:

1. Read the football.csv as a pandas dataframe
2. Find the number of teams participated in the football matches.
3. Sort the teams by the number of red cards they received and print.
4. Calculate the mean Red Cards of all the teams.
5. What is the minimum and maximum number of goals.
6. What is the mean shooting accuracy based on the number of goals scored.
7. What is the total number shots based on number of players used.

1. Read the football.csv as a pandas dataframe

```python
import pandas as pd
data=pd.read_csv('football.csv')
```

2. Find the number of teams participated in the football matches.

```python
teams=data['Team'].count()
print(teams)
```

3. Sort the teams by the number of red cards they received and print.

```python
data.sort_values(by='Red Cards',ascending=True)
```

4. Calculate the mean Red Cards of all the teams.

```python
data['Red Cards'].mean()
```

5. What is the minimum and maximum number of goals.

```python
max=data['Goals'].max()
print(max)
```

```python
min=data['Goals'].min()
print(min)
```

6. What is the mean shooting accuracy based on the number of goals scored.

```python
uni1=data.Goals.unique()
```

```python
for i in uni1:
    print("The mean shooting accuracy for goals scored",i,"is=",data[data['Goals']==i]["Shooting Accuracy"].mean())
```

7.What is the total number shots based on number of players used.

```python
uni2=data["Players Used"].unique()
```

```python
for i in uni2:
    print("The sum of shots based on number of teams",i,"is=",data[data['Players Used']==i]["Total shots (inc. Blocked)"].sum())
```

# Using the data ‘samp.user’ do the following:

1. Read the samp.user file as a dataframe with user_id field as index column (Hint: use read_table from pandas library)
2. What is the min and max age per occupation.
3. What is the mean age of all the persons based on occupation and gender.
4. For each occupation, calculate the percentage of women and men.

1. Read the samp.user file as a dataframe with user_id field as index column (Hint: use read_table from pandas library)

```python
data=pd.read_table('samp.user',delimiter='|')
```

```python
data
```

2. What is the min and max age per occupation.

```python
grp1=data.groupby(['occupation','age'])
```

```python
grp1['age'].first().groupby(['occupation']).max()
```

```python
grp1['age'].first().groupby(['occupation']).min()
```

3. What is the mean age of all the persons based on occupation and gender.

```python
grp2=data.groupby(['occupation','gender','age'])
```

```python
grp2['age'].first().groupby(['occupation','gender']).mean()
```

4. For each occupation, calculate the percentage of women and men.

```python
grp3=data.groupby(['occupation','gender'])
```

```python
((grp3['gender'].count())/(grp3['gender'].count().groupby(['occupation']).sum()))*100
```
